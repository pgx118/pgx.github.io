```python
import pandas as pd
```

```python

df=pd.read_excel("clear.xlsx")
df.describe()
```

```python
import matplotlib.pylab as plt 
plt.rcParams["font.sans-serif"]=["SimHei"]
plt.rcParams["axes.unicode_minus"]=False
plt.figure(figsize=(12,5))
plt.title("给区县的房子套数")
plt.xlabel("区县")
plt.ylabel("套数")
res=df["区县"].value_counts()
plt.bar(res.index,res.values)
for a,b in zip(res.index,res.values):
    plt.text(a,b,f"{b}套",ha="center",va="bottom",color="r")
plt.show()
```

```python
import matplotlib.pylab as plt 
plt.rcParams["font.sans-serif"]=["SimHei"]
plt.rcParams["axes.unicode_minus"]=False
plt.figure(figsize=(12,5))
plt.title("给区县的房子平均单价")
plt.xlabel("区县")
plt.ylabel("平均单价")
res=df.groupby("区县")["单"].mean().sort_values(ascending=False)
plt.bar(res.index,res.values)
for a,b in zip(res.index,res.values.round(2)):
    plt.text(a,b,f"{b}元/m²",ha="center",va="bottom",color="r")
plt.show()
```
