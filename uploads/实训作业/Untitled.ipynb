```python
def get_room_layout(df):
    # 筛选“户型”列为空的行（仅处理需要填充的数据，提升效率）
    data = df[df["户型"].isnull()]
    
    # 单层循环遍历需要填充的行（删除了无效的内层list循环）
    for index, row in data.iterrows():
        # 从标题提取户型并填充到当前行的“户型”列
        df.at[index, "户型"] = extra_layout(row["标题"])
    
    return df

# 1. 读取Excel数据（先确保文件路径正确、依赖安装完成）
df = pd.read_excel("长沙二手房.xlsx")

# 2. 调用函数填充户型空值
df = get_room_layout(df)

# 3. 查看结果
df.count()
```

```python
def get_direction(df):
    # 按“楼盘”分组，求“朝向”的众数，空则返回None
    res = df.groupby("楼盘")["朝向"].agg(lambda x: x.mode() if not x.mode().empty else None).to_dict()
    # 填充“朝向”为空的行
    for index, row in df[df["朝向"].isnull()].iterrows():
        df.at[index, "朝向"] = res[row["楼盘"]]
    return df

df = get_direction(df)
df.count()

# 检查仍为空的“朝向”行
df[df["朝向"].isnull()]  # 说明该方法无法填充，只能删除
df = df.dropna(subset=["朝向"])
df.count()
```

```python
# 填充“楼层”列为空的行，统一赋值为“别墅”
for index, row in df[df["楼层"].isnull()].iterrows():
    df.at[index, "楼层"] = "别墅"
df.count()
```

```python
df = df.dropna(subset=["户型","朝向","楼层","年份"])
df.isnull().sum()  # 6821行是完整的数据，没有缺失了
```

```python
import pandas as pd

# 方案1：先转字符串，再替换（推荐，兼容混合类型）
df["总价"] = df["总价"].apply(
    lambda x: float(str(x).replace("万", "")) if pd.notna(x) else x
)

# 方案2：更健壮的写法（增加类型判断，避免异常）
def convert_price(x):
    if pd.isna(x):  # 处理空值
        return x
    if isinstance(x, str):  # 仅对字符串做替换
        return float(x.replace("万", ""))
    elif isinstance(x, (int, float)):  # 数值类型直接返回
        return float(x)
    else:  # 其他类型默认返回空值
        return None

df["总价"] = df["总价"].apply(convert_price)

# 查看处理后的数据信息
df.info()
```

```python
df["单价"] = df["单价"].apply(lambda x:float(x.replace("元/㎡","")))
## 面积列清洗
df["面积"] = df["面积"].apply(lambda x:float(x.replace("㎡","")))
## 年份列清洗
df["年份"] = df["年份"].apply(lambda x:int(x.replace("年建","")))
```

```python
# 重命名列名，补充单位说明（总价单位：万；单价单位：元/㎡；面积单位：㎡；年份备注：年建）
df = df.rename(columns={"总价":"总价（万）","单价":"单价（元/㎡）","面积":"面积（㎡）","年份":"年份（年建）"})
# 查看重命名后的列名，验证修改结果
df.columns
```

```python
# 提取楼层属性（高/中/低/顶/底层）：匹配“高/中/低/顶/底+层”的格式
df["楼层属性"] = df["楼层"].str.extract(r"([低|高|中|顶|底]层)")

# 提取楼层数（数字）：匹配“共+数字+层”的格式，捕获其中的数字部分
df["楼层数"] = df["楼层"].str.extract(r"共(\d+)层")

# 查看提取后的前几行数据，验证特征提取结果
df.pop("楼层")
df.pop("详情页")
df.columns
```
